```python
def parse_csv(path):
    arr = []
    # Specify the path to your CSV file
    csv_file_path = path

    # Open and read the CSV file
    with open(csv_file_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        
        # Iterate through each row in the CSV file
        counter = 0
        repcount = 0
        repcountthree = 1
        for row in reader:
            tempcount = counter
            if counter >=17:
                reps = 17*repcountthree
                tempcount = counter - reps
            if tempcount == 5 or tempcount == 7 or tempcount == 9:
                repcount+=1
                newrow = convert_to_float(row)
                arr.append(newrow)
            if repcount == 3:
                repcountthree +=1
                repcount = 0
            counter+=1
    
    return arr

def convert_to_float(row):
    newrow = row
    row[1] = float(row[1])
    row[2] = float(row[2])
    return newrow
    



    
        
        
        
```

```python
dataset_right = []
for i in list_of_csv_right:
    dataset_right.append(parse_csv(i))
    

dataset_wrong = []
for i in list_of_csv_wrong:
    dataset_wrong.append(parse_csv(i))

print(dataset_right)
print(dataset_wrong)

    
    
```
