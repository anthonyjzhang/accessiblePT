```python
def parse_csv(path):
    arr = []
    # Specify the path to your CSV file
    csv_file_path = path

    # Open and read the CSV file
    with open(csv_file_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        
        # Iterate through each row in the CSV file
        counter = 0
        repcount = 0
        repcountfour = 1
        for row in reader:
            tempcount = counter
            if counter >=17:
                reps = 17*repcountfour
                tempcount = counter - reps
            if tempcount == 5 or tempcount == 7 or tempcount == 9 or tempcount ==11:
                repcount+=1
                newrow = convert_to_float(row)
                arr.append(newrow)
            if repcount == 4:
                repcountfour +=1
                repcount = 0
            counter+=1
    arr = np.array(arr)
    return arr

def convert_to_float(row):
    newrow = row
    row[1] = float(row[1])
    row[2] = float(row[2])
    return newrow
    
                
```

```python
def process_data_into_numpy(list_of_csv_right,list_of_csv_wrong):
    dataset_right = []
    for i in list_of_csv_right:
        temparr = parse_csv(i)
        dataset_right.append(temparr)
        

    dataset_wrong = []
    for i in list_of_csv_wrong:
        temparr = parse_csv(i)
        dataset_wrong.append(temparr)
        
        
    dataset_right = np.concatenate(dataset_right)
    dataset_wrong = np.concatenate(dataset_wrong)


    # Calculate the number of groups (x)
    x = dataset_wrong.shape[0] // 4
    # Reshape the data to (x, 4, 3)
    dataset_wrong = dataset_wrong[:x*4].reshape(x, 4, 3)

    # Calculate the number of groups (x)
    x = dataset_right.shape[0] // 4
    # Reshape the data to (x, 4, 3)
    dataset_right = dataset_right[:x*4].reshape(x, 4, 3)


    print(dataset_wrong)
    print(dataset_right)
    
process_data_into_numpy(list_of_csv_right,list_of_csv_wrong)
    
```
